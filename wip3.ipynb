```python
import numpy as np
import numba

fee = 0.003
p_sell = np.fromfile('btc_2024_minute_low.npy').astype(np.float32) * (1 - fee)
p_buy = np.fromfile('btc_2024_minute_high.npy').astype(np.float32) * (1 + fee)

# p_sell = (btc_train['low'].values * (1 - fee)).astype(np.float32)
# p_buy = (btc_train['high'].values * (1 + fee)).astype(np.float32)


@numba.jit
def minimize(p_buy, p_sell):
    n = p_buy.size
    arcs = np.zeros((n, 2, 2), dtype=np.float32)
    arcs[:,0,1] = -p_buy
    arcs[:,1,0] = p_sell
    arcs[0,1,:] = -np.inf
    arcs[-1,:,1] = -np.inf

    for i in range(n-1, 0, -1):
        max0 = max(arcs[i,0,0], arcs[i,0,1])
        max1 = max(arcs[i,1,0], arcs[i,1,1])
        arcs[i,0,:] -= max0
        arcs[i,1,:] -= max1
        arcs[i-1,:,0] += max0
        arcs[i-1,:,1] += max1
        # amax = arcs[i].max(-1) # numba error
        # arcs[i] -= amax[:, None]
        # arcs[i-1] += amax[None, :]

    max0 = max(arcs[0,0,0], arcs[0,0,1])
    arcs[0,0,:] -= max0
    arcs[0,1,:] = 0
    arcs[-1,:,1] = 0
    
    return arcs, max0

arcs, profit = minimize(p_buy, p_sell)
arcs, profit
```

```python
import torch

class Model(torch.nn.Module):
    def __init__(self, n=64):
        super().__init__()
        # self.n = n
        c = n
        self.drop = torch.nn.Dropout(p=0.1)
        Activation = torch.nn.LeakyReLU
        self.li = torch.nn.Sequential(
            torch.nn.Conv1d(2, c, n),
            Activation(),
            torch.nn.Conv1d(c, c, 2),
            Activation(),
            torch.nn.Conv1d(c, c, 2, dilation=2),
            Activation(),
            torch.nn.Conv1d(c, c, 2, dilation=4),
            Activation(),
            torch.nn.Conv1d(c, c, 2, dilation=8),
            Activation(),
            torch.nn.Conv1d(c, c, 2, dilation=16),
            Activation(),
            torch.nn.Conv1d(c, c, 2, dilation=32),
            Activation(),
            torch.nn.Conv1d(c, c, 2, dilation=64),
            Activation(),
            # torch.nn.Conv1d(c, c, 2, dilation=8),
            # torch.nn.Softplus(),
            # torch.nn.Conv1d(c, c, 2, dilation=4),
            # torch.nn.Softplus(),
            # torch.nn.Conv1d(c, c, 2, dilation=2),
            # torch.nn.Softplus(),
            torch.nn.Conv1d(c, 2, 2, dilation=1, bias=False),
        )

    def forward(self, x):
        n = x.shape[-1]
        x = x.log()
        x = x - x.mean()
        x = self.drop(x)
        x = self.li(x)
        return x

model = Model(256).cuda()
```

```python

fee = 0.003

p_sell = np.fromfile('btc_2024_minute_low.npy').astype(np.float32) * (1 - fee)
p_buy = np.fromfile('btc_2024_minute_high.npy').astype(np.float32) * (1 + fee)

# p_sell = (btc_train['low'].values * (1 - fee)).astype(np.float32)
# p_buy = (btc_train['high'].values * (1 + fee)).astype(np.float32)

p = torch.tensor([[p_sell, p_buy]]).cuda()

arcs, profit = minimize(p_buy, p_sell)
print('max profit', profit)
# print(arcs[0])
lossw = arcs.transpose(1,2,0)[None,...]
lossw = torch.tensor(lossw).cuda()
# print(lossw[..., 0])

lossw[0,0,0] *= 15
lossw[0,1,1] *= 15

with torch.no_grad():
    y = model(p)

    # y, _ = model(in_vecs=p.transpose(-2, -1))
    # y = y.transpose(-2, -1)

print(f'{p.shape=}')
print(f'{y.shape=}')
# print(f'{arcs.shape=}')
print(f'{lossw.shape=}')
```

```python
import numpy as np
from tqdm import tqdm

softplus = torch.nn.functional.softplus

opt = torch.optim.AdamW(
    model.parameters(),
    lr=1e-3,
    betas=(0.9, 0.995),
    weight_decay=1e-5,
)
model.train()

chunk = 32768
loss_avg = 0
loss_w = 0

for _ in (prog := tqdm(range(1600))):
    opt.zero_grad()

    # o = np.random.randint(0, p.shape[-1] - chunk)
    o = np.random.randint(0, chunk)
    # o = 0

    y = model(p[..., o:o+chunk])[..., :-1]

    # y, _ = model(in_vecs=p.transpose(-2, -1))
    # y = y.transpose(-2, -1)[..., :-1]

    ypos = softplus(y)
    yneg = softplus(-y)
    
    # offset = -0.25
    # ypos = softplus((y + offset), beta=2)
    # yneg = softplus(-(y + offset), beta=2)

    o2 = o + chunk - y.shape[-1]
    loss = ( ypos * lossw[:, :, 1, o2:o+chunk]
        + yneg * lossw[:, :, 0, o2:o+chunk] )

    # loss = -y * lossw[..., -y.shape[-1]:]
    # loss = loss.sum() / lossw.abs().sum()
    loss = -loss.sum() / lossw[..., o2:o+chunk].abs().sum()
    # loss = loss - arcs[-y.shape[-1]:, :, 0].sum()

    loss_w = min(loss_w + 1, 100)
    loss_avg += (loss.item() - loss_avg) / loss_w
    
    prog.set_description(f'{loss_avg:.5f}')
    loss.backward()
    opt.step()

loss_avg
```

```python
stock = 0
cash = 0
conf = 0.8
trades = 0
for i in range(y.shape[-1]):

    sell = (1 - y[1, i].item()) > conf
    buy = y[0, i].item() > conf

    if i == y.shape[-1] - 1:
        sell = 1
        buy = 0

    if stock==1 and sell:
        stock = 0
        cash += p_sell_test[i+1]
        trades += 1
    elif stock==0 and buy:
        stock = 1
        cash -= p_buy_test[i-1]
        trades += 1

cash, trades
```
